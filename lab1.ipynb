### Вероятностное пространство, формула Байеса

#### 1. Определить (с обоснованием), зависимы или независимы события

A и B - независимые:  
$$
P(A \cap B) = P(A) \cdot P(B)
$$

(a) Несовместные события - **зависимые**:

A и B - несовместные:  

$$
P(A \cap B) = 0
$$

Пусть P(A) > 0 и P(B) > 0, тогда:
$$
P(A) \cdot P(B) > 0
$$
$$
P(A \cap B) \ne P(A) \cdot P(B)
$$

(b) События, образующие σ-алгебру - **не дают информацию о зависимости, либо независимости**

Если $(A, B \in \Sigma)$, то  
- $(A^d \in \Sigma)$ (дополнение события A)
- $(A \cup B \in \Sigma)$  
- $(A \cap B \in \Sigma)$

Пример зависимости:

Пусть $A, B \in \Sigma$ и $A \subset B$, $P(A) > 0$ и $P(B) > 0$
$$
A \cap B = A
$$
$$
P(A \cap B) = P(A)
$$
Для независимости должно быть:
$$
P(A \cap B) = P(A) \cdot P(B)
$$
$$
P(A) = P(A) \cdot P(B)
$$
$$
1 = P(B)
$$

- Получаем $P(B) = 1$.  
- Если $P(B) \neq 1$, равенство не выполняется, значит $A$ и $B$ зависимы

Пример независимости:

Вероятностное пространство бросков двух монет. Пусть $A$ - выпадение орла на первой монете, $B$ - выпадение орла на второй монете. Множество всех возможных исходов:
$$
\Omega = \{О, Р\} \times \{О, Р\} = \{(О, О), (О, Р), (Р, О), (Р, Р)\}
$$

A и B образуют σ-алгебру, так как:
- $A^d = \{(Р, О), (Р, Р)\} \in \Sigma$
- $B^d = \{(О, Р), (Р, Р)\} \in \Sigma$
- $A \cup B = \{(О, О), (О, Р), (Р, О)\} \in \Sigma$
- $A \cap B = \{(О, О)\} \in \Sigma$
- Пересечения и объединения всех этих событий тоже входят в σ-алгебру

A и B независимы (см. пример независимости в пункте (c)).


(c) События, имеющие одинаковую вероятность - **не дают информацию о зависимости, либо независимости**

Пример зависимости:
Пусть A - бросок четного числа на игральной кости, B - бросок числа больше 3 на игральной кости.

$$
P(A) = P(B) = \frac{1}{2}
$$
$$
P(A \cap B) = P(\{4, 6\}) = \frac{1}{3} \ne P(A) \cdot P(B) = \frac{1}{4}
$$

Пример независимости:
Бросаем две монеты. Пусть A - выпадение орла на первой монете, B - выпадение орла на второй монете.

$$
P(A) = P(B) = \frac{1}{2}
$$
$$
P(A \cap B) = P(\{О, О\}) = \frac{1}{4} = P(A) \cdot P(B)
$$

Y - время прибытия второго парохода

X и Y равномерно распределены на отрезке [0, 24]

Рассмотрим квадрат 24 x 24, где ось X - время прибытия первого парохода, ось Y - время прибытия второго парохода

Общая площадь квадрата: $$S = 24 \cdot 24 = 576$$

Проведем линии X = Y - 2 и X = Y + 1.
Область, где одному из пароходов придется ждать, ограничена этими линиями и сторонами квадрата.

Найдем площадь этой области c помощью площадей треугольников, не входящих в область ожидания.:

### Случайный вектор и числовые характеристики

#### 1. Является ли функция плотностью распределения случайного вектора

$$
f_{\xi}(x,y) = \frac{e^{-2|y|}}{\pi (1+x^2)}
$$

Проверим два условия функции плотности распределения:
1. $f_{\xi}(x,y) \geq 0$ для всех $(x,y) \in R^2$: 

$e^{-2|y|}$ > 0 и $\pi (1+x^2) > 0$ для всех $x,y \in R^2$, значит $f_{\xi}(x,y) \geq 0$ для всех $(x,y) \in R^2$

2. $$\iint\limits_{R^2} f_{\xi}(x,y) dx dy = 1$$

$$
\iint\limits_{R^2} f_{\xi}(x,y) dx dy = \int\limits_{-\infty}^{\infty} \int\limits_{-\infty}^{\infty} \frac{e^{-2|y|}}{\pi (1+x^2)} dx dy = \int\limits_{-\infty}^{\infty} e^{-2|y|} dy \cdot \int\limits_{-\infty}^{\infty} \frac{1}{\pi (1+x^2)} dx
$$
$$
\int\limits_{-\infty}^{\infty} e^{-2|y|} dy = 2 \int\limits_{0}^{\infty} e^{-2y} dy = 2 \cdot \frac{-1}{2} e^{-2y} \Big|_{0}^{\infty} = 1
$$
$$
\int\limits_{-\infty}^{\infty} \frac{1}{\pi (1+x^2)} dx = \frac{1}{\pi} \cdot \arctan(x) \Big|_{-\infty}^{\infty} = \frac{1}{\pi} \cdot (\frac{\pi}{2} - (-\frac{\pi}{2})) = 1
$$

$$
\iint\limits_{R^2} f_{\xi}(x,y) dx dy = 1 \cdot 1 = 1
$$

Ответ: функция является плотностью распределения случайного вектора


#### 2. Совместное распределение случайных величин ξ и η задано таблицей:

Маргинальное распределение η:
$$
P_{\eta}(-1) = \frac{1}{8} + \frac{1}{3} = \frac{11}{24}

Сумма маргинальных распределений:

P_{\xi}(-1) + P_{\xi}(1) = \frac{1}{2} + \frac{1}{2} = 1
$$
$$
P_{\eta}(-1) + P_{\eta}(0) + P_{\eta}(1) = \frac{11}{24} + \frac{1}{4} + \frac{7}{24} = 1
$$

2. Вычислить математическое ожидание, ковариационную и корреляционную матрицы вектора (ξ, η)

Математическое ожидание:
$$
E[\xi] = -1 \cdot \frac{1}{2} + 1 \cdot \frac{1}{2} = 0
$$
$$
E[\xi^2] = (-1)^2 \cdot \frac{1}{2} + 1^2 \cdot \frac{1}{2} = 1
$$
$$
E[\eta] = -1 \cdot \frac{11}{24} + 0 \cdot \frac{1}{4} + 1 \cdot \frac{7}{24} = -\frac{4}{24} = -\frac{1}{6}
$$
$$
E[\eta^2] = (-1)^2 \cdot \frac{11}{24} + 0^2 \cdot \frac{1}{4} + 1^2 \cdot \frac{7}{24} = \frac{18}{24} = \frac{3}{4}
$$
$$
E[\xi \eta] = (-1)(-1) \cdot \frac{1}{8} + (-1)(0) \cdot \frac{1}{12} + (-1)(1) \cdot \frac{7}{24} + 1(-1) \cdot \frac{1}{3} + 1(0) \cdot \frac{1}{6} + 1(1) \cdot 0 = \frac{1}{8} - \frac{7}{24} - \frac{1}{3} = -\frac{1}{2}
$$

Дисперсия:
$$
D[\xi] = E[\xi^2] - (E[\xi])^2 = 1 - 0^2 = 1
$$
$$
D[\eta] = E[\eta^2] - (E[\eta])^2 = \frac{3}{4} - \left(-\frac{1}{6}\right)^2 = \frac{13}{18}
$$

Ковариационная матрица:
$$
cov(\xi, \eta) = E[\xi \eta] - E[\xi]E[\eta] = -\frac{1}{2} - 0 \cdot \left(-\frac{1}{6}\right) = -\frac{1}{2}
$$
$$
Cov = \begin{pmatrix}
1 & -\frac{1}{2} \\
-\frac{1}{2} & \frac{13}{18}
\end{pmatrix}
$$

Корреляционная матрица:
$$
\rho = \frac{cov(\xi, \eta)}{\sqrt{D[\xi] D[\eta]}} = \frac{-\frac{1}{2}}{\sqrt{1 \cdot \frac{13}{18}}} = -\frac{1}{2} \cdot \sqrt{\frac{18}{13}} = -\frac{3}{\sqrt{26}}
$$
$$
Corr = \begin{pmatrix}
1 & -\frac{3}{\sqrt{26}} \\
-\frac{3}{\sqrt{26}} & 1
\end{pmatrix}
$$

#### 4. Пусть $\xi \sim \text{U}_{-\pi, \pi}$ и $\eta_1 = \cos{\xi}, \eta_2 = \sin{\xi}$.

(a) Вычислить математическое ожидание, ковариационную и корреляционную матрицы вектора $(\eta_1, \eta_2)$

Плотность $\xi$:
$$
f_{\xi}(x) = \frac{1}{2\pi},\ x \in [-\pi, \pi]
$$

Математическое ожидание:
$$
E[\eta_1] = E[\cos{\xi}] = \int\limits_{-\pi}^{\pi} \cos{x} \cdot \frac{1}{2\pi} dx = 0
$$
$$
E[\eta_1^2] = E[\cos^2{\xi}] = \int\limits_{-\pi}^{\pi} \cos^2{x} \cdot \frac{1}{2\pi} dx
$$
$$
\int\limits_{-\pi}^{\pi} \cos^2{x} \cdot \frac{1}{2\pi} dx = \frac{1}{2\pi} \cdot \int\limits_{-\pi}^{\pi} \frac{1 + \cos{2x}}{2} dx = \frac{1}{4\pi} \cdot \left( x + \frac{\sin{2x}}{2} \right) \Big|_{-\pi}^{\pi} = \frac{1}{4\pi} \cdot (2\pi) = \frac{1}{2}
$$
$$
E[\eta_2] = E[\sin{\xi}] = \int\limits_{-\pi}^{\pi} \sin{x} \cdot \frac{1}{2\pi} dx = 0
$$
$$
E[\eta_2^2] = E[\sin^2{\xi}] = \int\limits_{-\pi}^{\pi} \sin^2{x} \cdot \frac{1}{2\pi} dx
$$
$$
\int\limits_{-\pi}^{\pi} \sin^2{x} \cdot \frac{1}{2\pi} dx = \frac{1}{2\pi} \cdot \int\limits_{-\pi}^{\pi} \frac{1 - \cos{2x}}{2} dx = \frac{1}{4\pi} \cdot \left( x - \frac{\sin{2x}}{2} \right) \Big|_{-\pi}^{\pi} = \frac{1}{4\pi} \cdot (2\pi) = \frac{1}{2}

$$
D[\eta_1] = E[\eta_1^2] - (E[\eta_1])^2 = \frac{1}{2} - 0^2 = \frac{1}{2}
$$
$$
D[\eta_2] = E[\eta_2^2] - (E[\eta_2])^2 = \frac{1}{2} - 0^2 = \frac{1}{2}
$$

Ковариационная матрица:
$$
cov(\eta_1, \eta_2) = E[\eta_1 \eta_2] - E[\eta_1]E[\eta_2] = E[\cos{\xi} \sin{\xi}] - 0 \cdot 0
$$
$$
E[\cos{\xi} \sin{\xi}] = \int\limits_{-\pi}^{\pi} \cos{x} \sin{x} \cdot \frac{1}{2\pi} dx = \frac{1}{2\pi} \cdot \int\limits_{-\pi}^{\pi} \frac{\sin{2x}}{2} dx = \frac{1}{4\pi} \cdot (-\frac{\cos{2x}}{2}) \Big|_{-\pi}^{\pi} = 0
$$
$$
cov(\eta_1, \eta_2) = 0
$$
$$
Cov = \begin{pmatrix}
\frac{1}{2} & 0 \\
0 & \frac{1}{2}
\end{pmatrix}
$$

Корреляционная матрица:
$$
\rho = \frac{cov(\eta_1, \eta_2)}{\sqrt{D[\eta_1] D[\eta_2]}} = \frac{0}{\sqrt{\frac{1}{2} \cdot \frac{1}{2}}} = 0
$$
$$
Corr = \begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
$$

Плотность распределения $\xi$:
$$
f_{\xi}(x) = \begin{cases}
2e^{-2x}, & x \geq 0 \\
0, & else
\end{cases}
$$

Плотность распределения $\eta$:
$$
f_{\eta}(y) = \begin{cases}
1, & y \in [0, 1] \\
0, & else
\end{cases}
$$

$\xi$ и $\eta$ независимы, значит плотность распределения суммы определяется сверткой:

$$
f_{\xi + \eta}(z) = \int\limits_{-\infty}^{\infty} f_{\xi}(x) f_{\eta}(z - x) dx
$$
$$
f_{\xi + \eta}(z) = \int\limits_{0}^{z} 2e^{-2x} \cdot 1 \, dx, \text{ при } z \in [0, 1]
$$
$$
f_{\xi + \eta}(z) = \int\limits_{z-1}^{z} 2e^{-2x} \cdot 1 \, dx, \text{ при } z > 1
$$
$$
f_{\xi + \eta}(z) = 0, \text{ при } z < 0
$$
$$
\int\limits_{0}^{z} 2e^{-2x} \, dx = \left[ -e^{-2x} \right]_{0}^{z} = 1 - e^{-2z}
$$
$$
\int\limits_{z-1}^{z} 2e^{-2x} \, dx = \left[ -e^{-2x} \right]_{z-1}^{z} = e^{-2(z-1)} - e^{-2z} = e^{2-2z} - e^{-2z}
$$
$$
f_{\xi + \eta}(z) = \begin{cases}
1 - e^{-2z}, & z \in [0, 1] \\
e^{2-2z} - e^{-2z}, & z > 1 \\
0, & else
\end{cases}
$$
